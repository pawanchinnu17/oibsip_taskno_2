```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

LOADING THE DATA

```python
df=pd.read_csv(r"C:\pawan chinnu data science\UNEMPLOYEMENT_ANALYSIS\Unemployment_Rate_upto_11_2020.csv")
df.head()
```

FINDING THE NULL VALUES

```python
print(df.isnull().sum())
```

```python
df.columns= ["State","Date","Frequency","Estimated Unemployment Rate","Estimated Employed","Estimated Labour Participation Rate","Region","longitude","latitude"]
```

```python
df.head()
```

DATA ANALYSIS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Select the columns for which you want to calculate correlations
selected_columns = ["Estimated Unemployment Rate",
                     "Estimated Employed",
                     "Estimated Labour Participation Rate",
                     "longitude",
                     "latitude"]

# Create a subset DataFrame with the selected columns
subset_df = df[selected_columns]

# Calculate the correlation matrix
correlation_matrix = subset_df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

DATA VISUALIZATION

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group data by "Region" and calculate the mean unemployment rate for each region
region_unemployment = df.groupby("Region")["Estimated Unemployment Rate"].mean().reset_index()

# Sort data by mean unemployment rate in descending order
sorted_region_unemployment = region_unemployment.sort_values(by="Estimated Unemployment Rate", ascending=False)

# Create a bar plot to visualize the regions with the highest unemployment rates
plt.figure(figsize=(10, 6))
plt.bar(sorted_region_unemployment["Region"], sorted_region_unemployment["Estimated Unemployment Rate"])
plt.xlabel("Region")
plt.ylabel("Mean Unemployment Rate")
plt.title("Regions with Highest Unemployment Rates")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Group data by "Region" and calculate the mean unemployment rate for each region
region_unemployment = df.groupby("Region")["Estimated Unemployment Rate"].mean().reset_index()

# Sort data by mean unemployment rate in descending order
sorted_region_unemployment = region_unemployment.sort_values(by="Estimated Unemployment Rate", ascending=False)

# Create a professional-looking bar plot using Seaborn
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")  # Set the style

# Define a custom color palette for the bars
custom_palette = sns.color_palette("coolwarm", n_colors=len(sorted_region_unemployment))

# Create the bar plot
ax = sns.barplot(x="Region", y="Estimated Unemployment Rate", data=sorted_region_unemployment, palette=custom_palette)

# Add data labels above the bars
for p in ax.patches:
    ax.annotate(f"{p.get_height():.2f}", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha="center", va="center", fontsize=12, color="black", xytext=(0, 5),
                textcoords="offset points")

# Customize plot appearance
plt.xlabel("Region", fontsize=14, labelpad=15)
plt.ylabel("Mean Unemployment Rate", fontsize=14, labelpad=15)
plt.title("Regions with Highest Unemployment Rates", fontsize=16, pad=20)
plt.xticks(rotation=45, ha="right", fontsize=12)

# Remove top and right spines
sns.despine(top=True, right=True)

# Adjust y-axis labels
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```

DIFFERENT TYPES OF PLOTS REGARDING THE Estimated Unemployment Rate

BOX PLOT

```python

plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Customize the appearance of the boxplot
sns.boxplot(x="State", y="Estimated Unemployment Rate", data=df, palette="coolwarm")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha="right", fontsize=12)

# Add labels and title
plt.xlabel("State", fontsize=14, labelpad=15)
plt.ylabel("Unemployment Rate", fontsize=14, labelpad=15)
plt.title("Unemployment Rate by State (Boxplot)", fontsize=16, pad=20)

plt.tight_layout()
plt.show()
```

VIOLIN PLOT

```python

plt.figure(figsize=(12, 6))

# Customize the appearance of the violin plot
sns.violinplot(x="State", y="Estimated Unemployment Rate", data=df, palette="coolwarm")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha="right", fontsize=12)

# Add labels and title
plt.xlabel("State", fontsize=14, labelpad=15)
plt.ylabel("Unemployment Rate", fontsize=14, labelpad=15)
plt.title("Unemployment Rate by State (Violin Plot)", fontsize=16, pad=20)

plt.tight_layout()
plt.show()
```

SWARMPLOT

```python

plt.figure(figsize=(12, 6))

sns.swarmplot(x="State", y="Estimated Unemployment Rate", data=df, palette="coolwarm", s=5)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha="right", fontsize=12)

# Add labels and title
plt.xlabel("State", fontsize=14, labelpad=15)
plt.ylabel("Unemployment Rate", fontsize=14, labelpad=15)
plt.title("Unemployment Rate by State (Swarm Plot)", fontsize=16, pad=20)

plt.tight_layout()
plt.show()
```

Estimated Unemployment Rate BASED ON THECOMBINATION OF REGION AND STATES

```python
import pandas as pd
import matplotlib.pyplot as plt

df['Date'] = pd.to_datetime(df['Date'])
regions = df['Region'].unique()

# Create a line plot for each region
plt.figure(figsize=(12, 6))
for region in regions:
    region_data = df[df['Region'] == region]
    plt.plot(region_data['Date'], region_data['Estimated Unemployment Rate'], marker='o', linestyle='-', label=region)

# Customize plot appearance
plt.title('Unemployment Rate by Date and Region', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Estimated Unemployment Rate', fontsize=14)
plt.grid(True)
plt.legend(loc='upper left', fontsize=12)

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```
